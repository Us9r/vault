kenapa kecamatan sukarami malah bertambah

```python
# prompt: deteksi perubahan toko yang kecamatan sukarami

# Filter data untuk kecamatan Sukarami sebelum dan sesudah perbaikan Ilir Timur III
df_sukarami_before = df_indomaret_to_repair[df_indomaret_to_repair['kecamatan'] == 'Sukarami']
df_sukarami_after = df_indomaret_to_repair_ilir_timur_3[df_indomaret_to_repair_ilir_timur_3['kecamatan'] == 'Sukarami']

# Identifikasi toko yang ada di Sukarami setelah tetapi tidak ada sebelum
toko_baru_di_sukarami = df_sukarami_after[~df_sukarami_after['title'].isin(df_sukarami_before['title'])]

print("\nToko Indomaret baru yang muncul di Kecamatan Sukarami setelah perbaikan Ilir Timur III:")
print(toko_baru_di_sukarami[['title', 'neighborhood', 'kecamatan']])

# Identifikasi toko yang ada di Sukarami sebelum tetapi tidak ada setelah
toko_hilang_dari_sukarami = df_sukarami_before[~df_sukarami_before['title'].isin(df_sukarami_after['title'])]

print("\nToko Indomaret yang hilang dari Kecamatan Sukarami setelah perbaikan Ilir Timur III:")
print(toko_hilang_dari_sukarami[['title', 'neighborhood', 'kecamatan']])

# Bandingkan index dari toko yang ada di Sukarami sebelum dan setelah
# Ini bisa membantu melihat apakah ada toko yang diubah kecamatannya menjadi Sukarami
sukarami_index_before = set(df_sukarami_before.index)
sukarami_index_after = set(df_sukarami_after.index)

# Index yang baru muncul di Sukarami
newly_sukarami_indices = list(sukarami_index_after - sukarami_index_before)

if newly_sukarami_indices:
    print(f"\nIndex toko yang sekarang berstatus di Sukarami (sebelumnya bukan): {newly_sukarami_indices}")
    print("Detail toko yang berpindah ke Sukarami:")
    print(df_indomaret_to_repair_ilir_timur_3.loc[newly_sukarami_indices, ['title', 'neighborhood', 'kecamatan']])
else:
    print("\nTidak ada toko dengan index baru yang berstatus di Sukarami.")

# Index yang hilang dari Sukarami
missing_from_sukarami_indices = list(sukarami_index_before - sukarami_index_after)

if missing_from_sukarami_indices:
    print(f"\nIndex toko yang sebelumnya di Sukarami (sekarang bukan): {missing_from_sukarami_indices}")
    # Untuk melihat status toko ini sekarang, kita bisa cari di df_indomaret_to_repair_ilir_timur_3
    print("Status toko yang keluar dari Sukarami:")
    print(df_indomaret_to_repair_ilir_timur_3.loc[missing_from_sukarami_indices, ['title', 'neighborhood', 'kecamatan']])
else:
    print("\nTidak ada toko yang keluar dari Sukarami.")

# Cek data asli sebelum perbaikan Ilir Timur III untuk index yang sekarang di Sukarami
if newly_sukarami_indices:
    print("\nStatus toko yang sekarang di Sukarami (sebelum perbaikan Ilir Timur III):")
    print(df_indomaret_to_repair.loc[newly_sukarami_indices, ['title', 'neighborhood', 'kecamatan']])

```

setelah dicek ternyata lokasi (fake)

```python
# prompt: toko indomaret di kecamatan sukarami yang tidak memiliki substring Kebun Bunga atau Sukabangun atau Sukajaya atau Sukarami atau Sukodadi atau Talang Betutu atau Talang Jambe di neighborhood

# Filter toko di kecamatan Sukarami
df_sukarami = df_indomaret_to_repair_ilir_timur_3[df_indomaret_to_repair_ilir_timur_3['kecamatan'] == 'Sukarami']

# Daftar substring yang tidak diinginkan di 'neighborhood'
unwanted_substrings = ['Kebun Bunga', 'Sukabangun', 'Sukajaya', 'Sukarami', 'Sukodadi', 'Talang Betutu', 'Talang Jambe']

# Filter toko yang neighborhood-nya tidak mengandung substring yang tidak diinginkan
# Gunakan regex untuk memastikan tidak ada substring yang cocok
# ^((?!Kebun Bunga|Sukabangun|Sukajaya|Sukarami|Sukodadi|Talang Betutu|Talang Jambe).)*$
# Ini artinya baris harus cocok dengan ekspresi regular:
# ^ : Memulai baris
# ((?!substring1|substring2|...).)* : Ini adalah negative lookahead assertion.
# (?!substring1|...) : Memastikan bahwa substring apapun dalam kurung tidak muncul di posisi saat ini.
# . : Mencocokkan karakter apapun (kecuali baris baru).
# * : Mencocokkan karakter sebelumnya nol atau lebih kali.
# )* : Mengulangi grup (yaitu, mencari karakter apapun yang tidak diikuti oleh substring yang tidak diinginkan) nol atau lebih kali.
# $ : Mengakhiri baris
# na=False : Memastikan bahwa nilai NaN tidak dipertimbangkan dalam pencarian substring
# case=False : Mengabaikan perbedaan huruf besar/kecil
df_filtered = df_sukarami[~df_sukarami['neighborhood'].str.contains('|'.join(unwanted_substrings), case=False, na=False)]

print("\nToko Indomaret di Kecamatan Sukarami tanpa substring tertentu di Neighborhood:")
print(df_filtered[['title', 'neighborhood', 'kecamatan']])
print(f"\nJumlah toko: {len(df_filtered)}")
```

```python
# prompt: visualisasi kota palembang dengan df_final dengan folium

!pip install folium geopandas
import folium
import geopandas as gpd
from shapely.geometry import Point

# Koordinat pusat Kota Palembang

avg_lat = df_final['location/lat'].mean()
avg_lng = df_final['location/lng'].mean()

# Create a map centered around Palembang
m = folium.Map(location=[avg_lat, avg_lng], zoom_start=13)

# Tambahkan marker untuk setiap toko Indomaret
for index, row in df_final.iterrows():
    # Pastikan nilai latitude dan longitude tidak NaN
    if pd.notna(row['location/lat']) and pd.notna(row['location/lng']):
        folium.Marker(
            location=[row['location/lat'], row['location/lng']],
            popup=f"<b>{row['title']}</b><br>Kecamatan: {row['kecamatan']}<br>Rating: {row['totalScore']}<br>Reviews: {row['reviewsCount']}",
            icon=folium.Icon(color='blue', icon='info-sign')
        ).add_to(m)

# Tampilkan peta
m
```
